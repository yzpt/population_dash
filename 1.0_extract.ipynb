```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
gdf = gdf[gdf['departement'] == '59']
gdf['pop'] = gdf['p21_pop']
```

```python
fig = go.Figure()

fig.add_trace(
    go.Choroplethmapbox(
        geojson=gdf.__geo_interface__,
        locations=gdf.index,         # The feature's array of locations
        z=gdf['pop'],       # The values you want to display in the choropleth
        colorscale="Viridis",        # Choose your color scale
        # ajdust color range
        zmin=0,
        zmax=100000,
        marker_opacity=0.5,
        marker_line_width=0,
        showlegend=False,
        text=gdf['nom'],
        hoverinfo='text+z',
    )
)

# Set up the map layout
fig.update_layout(
    mapbox_style="carto-darkmatter",
    mapbox_zoom=5,              # Set zoom level
    mapbox_center={"lat": 50.5, "lon": 2.8},  # Center on the map
    margin={"r":0,"t":0,"l":0,"b":0}  # Remove margins
)

fig.show()
```
